**Box Plots (Visualization of Data)**

This plots is mainly used tod etect the Outliers

In Box plot we need to find the Upper Whisker and Lower Whiskers.

Upper Whisker - Maxima range

Lower Whisker --Minima range

Data that lies beyond teh upper Whisker and Lower Whisker are considered as Outliers.

 **Lower Whisker =Q1-(1.5*IQR)**

**Upper Whisker =Q3+(1.5*IQR) **

**IQR =Q3 - Q1**

Q3=75th  Percentile

Q1 =25th  percentile

median =Q2 =50th percentile
