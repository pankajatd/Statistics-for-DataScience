**Inferential Statistics**

 In Inferential Statistics whatever  "Results'  we get for the Sample data,what is the confidence level to apply teh same Results for the "Population" data.? This is where the concept of Hypothesis testing comes into picture.

 Inferential Statistics is about Predicting the Population results based on the results obtained from the Sample Data.

 For this we use the Hypothesis testing.

Emperical Rule:

When the Data belongs to the Normal Distribution we have the below Emperical rule:

(mu+sigma,mu-sigma) --68 % of the Data lies in this range

(mu+2sigma,mu-2sigma) --95 % of the Data lies in this range

(mu+3sigma,mu-3sigma) --99.7 % of the Data lies in this range


Data beyond (mu+3sigma,mu-3sigma) is an Outlier.

Correlation Coefficients: Correlation coeefiecients lie between +1 to -1

1.Features are either Positively Correlated -- If feature f1 is increasing then feature f2 is also increasing.
If the value is approximately around 0.85 ..This means features are highly correlated.
Can use either one of the feature

2.Features are negatively correlated.--If feature f1 is increasing then feature f2 is decreasing.
If the value is approximately around -0.85 ..This means features are highly correlated.
Can use either one of the feature

3.Correlation is Zero --This means there is no Correlation between the Features.

If the correlation value between the two features f1 and f2 is 0,then  we need to use both the features.


This is used in **" Feature Selection"**


If F1 and F2 are highly correlated means that both of them give the same Information. --Hence one of the feature can be removed.

Hence this is used as the **"Feature Selection Technique"**
