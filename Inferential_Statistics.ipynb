**Hypothesis Testing** is a part of  Inferential statistics


Hypothesis Testing are of two types:

Null Hypothesis --- Assumes that x_bar = mu --NULL Hypothesis is Doubtful condition

Alternate Hypothesis --Assumes that x_bar!=mu --- Alternate Hypothesis is one which we believe.



Multiple statistical Tests are conducted:

1. T-test
2. ANOVA Test
3. Z-test
4.Chi-Square Test

Based on these Tests conducted ,we either 1. "Reject the Null Hypothesis
                                     
                                          2. Fail to Reject teh Null Hypothesis

P-Value --P-value is teh Probability of Obtaining the results

 i. It is in Favour of Null Hypothesis.


 ii. If P-value is higher then we **Fail to Reject the Null Hypothesis** H0

 iii. If P-Value is lower then we **Reject the Null Hypothesis**




**Problem Statement:**

The average height of an Adults in a certain country is known to be 168cm.A scientist believes that the average height of adult residents in a certain city in that country  is different.
The scientist measures the average height of 40 randomly selected adult residents of teh city and finds that the average height is 171cm with a standard deviation of 7cm.
Test the scientist's claim using a significance level of 0.10.


From the above problem statement we have the following information:

1. population mean(mu)= 168

2. sample mean(x_bar) =171

3.Sample size (n) =40

4.Sample Std Deviation (s) =7


From this we have  

Null Hypothesis H0: mu=168

Alternate Hypothesis H1: mu!=168


Assumption and steps to solve the above propblem:


1. usually the problem related with height of a person and age are a Normal Distribution.


2. Check if Population Std Deviation is Give?
   i. Yes(Population STD Deviation is given) -- If Yes check the Number of data points in the sample---> If n>30  go with the z-test
                                 If n<3o go with the T-test


   ii. No(Population STD Deviation is not given)-- Go with the T-test


   Formula for T-test and Z-test


   T-test = (x_bar-mu)/(s/sqrt(n))

   Z-test = (x_bar-mu)/(sigma/sqrt(n))


   For the above problem statement we  apply T-test as we donot have Population Std Deviation Sigma


   T-test = (171-168)/(7/sqrt(40))


   **Note : T-test and Z-test signify the "Distribution of the Data"**
   If teh Data is Normal or Symmetric we go witrh t-Test or Z-Test type of Satistical tests.

   T value we get from T-test is compared with the Critical value ,the Critical value is obtained from the table for the given alpha value
    In the above problem statement alpha =0.10

    Critcal value =1.68

    Tvalue =1.21

    If Critical value(P-value)  > Tvalue --> then we  Reject the NULL hpoyhesis



```python

mu= 168
x_bar =171
n =40
s =7
```

```python
import numpy as np
T_test = (171-168)/(7/np.sqrt(40))
print(f"T-test value is: {T_test}")
```

```python
P-value =1.68

T_test =2.71

#Statistical_value < P_value(Critical value)  --> Fail to Reject the Null Hypothesis

#Statistical_value > P_value(Critical value)  --> Reject the Null Hypothesis

#hence
1.68 <2.71 #we Reject teh NULL Hypothesis
```
