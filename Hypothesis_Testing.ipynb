Hypothesis Testing
Hypothesis Tesing has two parts
1. **Null Hypothesis H0-**--Null hypothesis is a condition that is doubtfull
2. **Alternate Hypothesis Ha**-Alternate Hypothesis is a condition where we certainally believe


There can be two things that canhappen

1. Reject the Null hypothesis
2. fail to Reject the Null hypothesis.

Example of Hypothesis testing

Null hypothesis -- Drinking coffee has no linkage to Heart attack.

Alternate Hypothesis- -Drinking coffe increases the chances of Heart attack.

To reject the Null hypothesis is to conclude that it is false.

To accept the Null Hypothesis does not mean that it is True.--it just indicates that we donot have enough evidence to believe otherwise.


Type I error --(False Negative)(denoted by Alpha) Actually Null Hypothesis is true and You still Reject the Null Hypothesis.


Correct--- (True Positive) Actually Null Hypothesis is False and you Reject th Null Hypothesis


 Correct---- (True Negative)Actually Null hypothesis is True and You Reject to Fail the Null Hypothesis

Type II error --(False Positive)(denoted by Beta) Actually Null Hypothesis is False and You still Fail to Reject the Null Hypothesis.

Alpha abd Beta are called the Level of Significance

Our aim should be always to reduce the value of Alpha and Beta.
