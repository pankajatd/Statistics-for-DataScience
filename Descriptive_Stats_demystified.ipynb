**Variance** - It is the overall spread of the Dataset w.r.t to the Mean values.

sigma**2 (Population Variance) --mu --> Indicates the Population mean.

In Population variance we divide the Nr/ N

Population variance = Summation(1->N(xi-mu)**2/N


S**2(Sample Square) ---x**bar --> Indicates the sample mean
In Sample Variance we divide the Nr / n-1

Sample variance = Summation(1->n(xi-x**bar)**2/n-1


We see that in both Population and Sample Variance the difference of each value is found w.r.t to mean.

mu(Population Mean)=summation(1 to N(xi)/N

x**bar(Sample Mean)=summation(1 to n(xi)/n-1

Why it is N in the denominator for the STD Deviation for Population Data and (n-1) for the STD Deviation for Sample Data?

A..We know that

x_bar approximately equal to mu (Mean for Sample and Population respectively)

S approximately equal to Sigma (STD Deviation  for Sample and Sigma is STD Deviation for Population respectively)

Now when the denominator for Sample STD  is taken as n instead of (n-1),it was obeserved that there is a significant difference in the Sample STD_Deviation and Population STD_Deviation,which is not correct.

Hence to ahandle this we take (n-1) for the Sample STD_Deviation instead of n.

with (n-1) we see that STD_Deviation_Samples and STD_Deviation_Population is approximately equal.

With this the Data is Unbiased.
