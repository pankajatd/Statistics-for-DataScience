**Measures of Dispersion : Range and IQR**

Range -

The spread of the data can be measured by using the **'Range'**

Range is the difference in the Max value in the data and Min value in the data

The Max and Min value from the data can be obtained by using teh describe () function .. dataset.describe() ...gives the max and min values for each column.

Range =(Max - Min)

Range is very easy to calculate , but it is not teh robust method when there are extreme values or Outliers.



Range explaination with an Example

```python
sal=[34500,30700,32900,36000,34100,33800,32500]
```

```python
min_sal = min(sal)
print(min_sal)
```

```python
max_sal = max(sal)
print(max_sal)
```

```python
Range = max_sal - min_sal
print(Range)
```
