Population Variance = 1/N(Summation(xi-mu)**2 where i ranges from 1 to N

here mu -> Population Mean

Sample Variance = 1/n-1(Summation(xi-x_bar)**2 where i ranges from 1 to n

x_bar is the Sample Mean

Interview Question:

In Sample variance why we divide the Denominator by (n-1) instead of n ??

Ans: 1. When we divide thesample varience by (n-1) we get teh higher variance value,as compared to the variance value when we divide by n.

2. From the above obeservation of (Point 1) we can say that Sample Variance is Inderectly propotional to n-1.

Conclusion : Lower the value of n ,higher is the value of Sample Variance.
Higher value of Sample Variance ,would approximately be equal to the Population variance.
Hence the instead of divinding by n,we divide Sample variaanc eby (n-1)

 **When we are dividing the Sample variance by (n-1) it is giving the
  UnBiased Estimator w.r.t to the Population Variance as we get the higher Sample Variance**


 **When we are dividing the Sample variance by (n) it is giving us the
  Biased Estimator w.r.t to the Population Variance as we get the Lower Sample Variance**


  **Since we aim to get the UnBiased Estimator we divide the sample variance by (n-1)**
