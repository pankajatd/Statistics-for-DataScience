If we get a Bell shaped Curme or Normal Curve then it is very easy identify the outliers in the Data  .

if data point < mu-3*sigma and data point > m+3*sigma the n that data point is an Outlier.

To find the outliers in this Normal Curve , we need to find if there any Datapoints that lie outside (mu-3*sigma,mu+3*sigma)


To find the Outliers ,we need to first find mu and Std Deviation

```python
mean_height=dataset.Height.mean()
print(mean_height)
```

```python
std_dev_height=dataset.Height.std()
print(std_dev_height)
```

Now calculate mu - 3 *sigma #where sigma is the STD Deviation

mu =-> mean_height

```python
mean_height - 3*std_dev_height
```

```python
mean_height + 3*std_dev_height
```

From the above values we caan find that if the Dataset contains values below 54.82 and dataset contains value above 77.91 , then those points are Outliers

Use the below calculations to find the Outliers

```python
dataset[(dataset.Height <(mean_height - 3  * std_dev_height)) | (dataset.Height >(mean_height +3 * std_dev_height))]
```

We saw that above records are the Outliers .

To remove those Outliers an dget the Dataset without Outlieres

```python
data_No_Outliers = dataset[(dataset.Height > (mean_height - 3 * std_dev_height)) & (dataset.Height < (mean_height+ 3 * std_dev_height))]
```
