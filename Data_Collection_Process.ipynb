**Data Collection process**


1. Specify the Objective of the Study
Ex: Study how many covid cases are there in Delhi

2. Identify the variable(s) of Interest
Ex: a. Numerical data(quantitative)-Take on Numerical values(add ,subtract,average etc)
     i. Numerical has **Continious** type of Data - Take on an infinite number of values within some given range(Ex: Amount of Water  in a Container)
     ii. Numerical has **Discerete** type of Data - Take on specific setof Numerical values(numof Subscribers)

     b. Categorical(qualitative)- Take on limited distinct categories that can be identified with Numbers(but cannot do any mathematical operation)

     i. Nominal - Where there is No specific order for the category
     Ex : Gender

     ii. Ordinal -where there is a specific order for the Category
     Ex: Level of Education(Phd,Masters,Bachelors)
