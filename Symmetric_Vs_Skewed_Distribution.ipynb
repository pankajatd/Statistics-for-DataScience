Measure of Skweness

Symmetric Curve Or Normal Curve: In a Symmettric Curve Mean,Median and Mode are all Equal

i.e Mean=Median=Mode

Left/Negatively Skewed-

If a distribution has a long Left tail,it is Left Skewed(i.e Mean<Median<Mode)

Example : Retirement Age(Chances of People with less age getting retired is Less) , Life Cycle of Human being(Chances of people with with less age dying is less)



Right/Positively Skewed-

If a distribution has a long Right tail,it is Left Skewed(i.e Mean>Median>Mode)

Example :Salary of an Employee working in an organization( Employees of certain Age group earn same range salary, few exception that some employees earn very higher range of Salary)
