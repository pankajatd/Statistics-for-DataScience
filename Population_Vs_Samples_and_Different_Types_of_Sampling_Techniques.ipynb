**Sample:** -All statistical Analysis cannot be done on the huge population data ,hence we take the sample from the Population Data.

1. Sample is the Subset of the Population that we have.

2. Sample is set of measurements that is collected from the Population.

3. Infrences that we get for the Sample ,we assume that applies to the Population data.


**Types of Sampling techinques:**

1.**Simple random Sampling- **Randomly picking the Sample Data from the Population

2.**Stratified Sampling -** Population data is made into separate groups and then sample data is picked from each group. that is called as Homogenous strata. The Group or the strat should have common property  between thm.
From this Homogenous strate pick the Data randomly.

3.**Cluster Sampling: **Here we make multiple cluster of the Population Data that have common Properties. And pick randmoly multiple clusters.


4.**Systematic Sampling-** Here the population data is picked in a Systematic manner.

Ex : Pick every 4th element from the Population Data.
